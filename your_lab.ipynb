```python
# Remove non-numeric characters and convert the column to float
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Check if there are any NaN values after conversion
print(df['discounted_price'].isnull().sum())
```

```python
# Option 1: Fill NaN values with a specific value, like 0
df['discounted_price'].fillna(0, inplace=True)

# Option 2: Drop rows with NaN values
# df.dropna(subset=['discounted_price'], inplace=True)
```

```python
# Calculate the first (Q1) and third (Q3) quartiles
Q1 = df['discounted_price'].quantile(0.25)
Q3 = df['discounted_price'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_clean = df[(df['discounted_price'] >= lower_bound) & (df['discounted_price'] <= upper_bound)]

# Verify the changes
print(df_clean.head())
```

```python
# Extract the outliers that were removed
outliers = df[(df['discounted_price'] < lower_bound) | (df['discounted_price'] > upper_bound)]

# Display the outliers
print(outliers)
```

```python
import pandas as pd

# 1. Load the dataset (assuming you have a CSV file)
df = pd.read_csv('your_dataset.csv')  # replace with your actual file path

# Display the dataset before any reduction
print("------ Dataset Before Data Reduction ------")
print(df.head())  # Display the first 5 rows of the original dataset

# 2. Sampling (Take a 10% sample of the data)
df_sample = df.sample(frac=0.1, random_state=42)  # 10% random sample of the data

# Display the dataset after sampling
print("\n------ Dataset After Sampling (10%) ------")
print(df_sample.head())  # Display first 5 rows of the sampled data

# 3. Dropping Unnecessary Columns
# (For example, we'll drop 'product_link', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link')
df_reduced = df.drop(columns=['product_link', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link'])

# Display the dataset after dropping unnecessary columns
print("\n------ Dataset After Dropping Unnecessary Columns ------")
print(df_reduced.head())  # Display first 5 rows of the reduced dataset

```

```python
min_value = df['discounted_price'].min()
max_value = df['discounted_price'].max()
mean_value = df['discounted_price'].mean()
median_value = df['discounted_price'].median()
mode_value = df['discounted_price'].mode()[0]

print(f'Min: {min_value}, Max: {max_value}, Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}')
```

```python
# 4.3 Dispersion Measures

# Calculate Range
range_value = df['discounted_price'].max() - df['discounted_price'].min()

# Calculate Quartiles
Q1 = df['discounted_price'].quantile(0.25)
Q3 = df['discounted_price'].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Calculate Variance
variance = df['discounted_price'].var()

# Calculate Standard Deviation
std_dev = df['discounted_price'].std()

# Print the results
print(f'Range: {range_value}, Q1: {Q1}, Q3: {Q3}, IQR: {IQR}, Variance: {variance}, Std Dev: {std_dev}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
